```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
import re
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from sklearn.model_selection import cross_validate
from sklearn.ensemble import RandomForestClassifier
from sklearn import svm
from sklearn.linear_model import LogisticRegression
```

Data

```python
# Read the CSV file
df = pd.read_csv('../../data/new_data.csv')

# Extract all messages from the "Message" column
data = df['Message'].tolist()
# save data to txt
with open('data.txt', 'w', encoding='utf-8') as f:
    for item in data:
        f.write("%s\n" % item)

ground_truth = df['R2DiscussionType'].tolist()

# Print all messages
# ground_truth = set(ground_truth)
print(ground_truth)
```

```python
# preprocess the data

preprocessed_data = []

def preprocess_string(text_data):
    text_data = text_data.lower()

    # Remove punctuation
    text_data = re.sub(r'[^\w\s]', '', text_data)

    # Tokenize the text
    tokens = word_tokenize(text_data)

    # Remove stopwords
    stop_words = set(stopwords.words('english'))
    filtered_tokens = [word for word in tokens if word not in stop_words]

    # Lemmatization
    lemmatizer = WordNetLemmatizer()
    lemmatized_tokens = [lemmatizer.lemmatize(word) for word in filtered_tokens]

    # Reconstruct the preprocessed text
    preprocessed_text = ' '.join(lemmatized_tokens)

    return preprocessed_text


# Sample text data
for text_data in data:
    preprocessed_text = preprocess_string(text_data)
    # Append the preprocessed text to the list
    preprocessed_data.append(preprocessed_text)
    print(preprocessed_text)

```

```python
# Initialize TF-IDF Vectorizer
tfidf_vectorizer = TfidfVectorizer()

# Fit and transform the data
tfidf_matrix = tfidf_vectorizer.fit_transform(preprocessed_data)

# Get the feature names (words)
feature_names = tfidf_vectorizer.get_feature_names_out()

# Convert to dense matrix
tfidf_matrix_dense = tfidf_matrix.toarray()

# Print the TF-IDF matrix
print(tfidf_matrix_dense)

# save to file txt
with open('tfidf_matrix_dense.txt', 'w') as f:
    for item in tfidf_matrix_dense:
        f.write("%s\n" % item)  
```

```python
# Perform cross-validation
logistic_classifier = LogisticRegression()
logistic_scores = cross_validate(logistic_classifier, tfidf_matrix_dense, ground_truth, cv=4, scoring='accuracy')
logistic_scores = logistic_scores['test_score']
print("%0.2f accuracy with a standard deviation of %0.2f" % (logistic_scores.mean(), logistic_scores.std()))

svm_classifier = svm.SVC()
svm_scores = cross_validate(svm_classifier, tfidf_matrix_dense, ground_truth, cv=4, scoring='accuracy')
svm_scores = svm_scores['test_score']
print("%0.2f accuracy with a standard deviation of %0.2f" % (svm_scores.mean(), svm_scores.std()))

rndForest_classifier = RandomForestClassifier()
rndForest_scores = cross_validate(rndForest_classifier, tfidf_matrix_dense, ground_truth, cv=4, scoring='accuracy')
rndForest_scores = rndForest_scores['test_score']
print("%0.2f accuracy with a standard deviation of %0.2f" % (rndForest_scores.mean(), rndForest_scores.std()))
```

```python
# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(tfidf_matrix_dense, ground_truth, test_size=0.2, random_state=42)

# Initialize and train the logistic_classifier
logistic_classifier = LogisticRegression()
logistic_classifier.fit(X_train, y_train)

# Predictions
predictions = logistic_classifier.predict(X_test)

# Evaluate the model
accuracy = logistic_classifier.score(X_test, y_test)
print("Accuracy:", accuracy)
```

```python


# Initialize and train the classifier
svm_classifier = svm.SVC()
svm_classifier.fit(X_train, y_train)

# Predictions
predictions = svm_classifier.predict(X_test)

# Evaluate the model
accuracy = svm_classifier.score(X_test, y_test)
print("Accuracy:", accuracy)
```

```python


# Initialize and train the classifier
rndForest_classifier = RandomForestClassifier()
rndForest_classifier.fit(X_train, y_train)

# Predictions
predictions = rndForest_classifier.predict(X_test)

# Evaluate the model
accuracy = rndForest_classifier.score(X_test, y_test)
print("Accuracy:", accuracy)
```
